```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from abc import ABCMeta, abstractmethod 
```

```python
class segment:
    """
    A segment of a fractal tree
    """
    __metaclass__ = ABCMeta
    
    def __init__(self, parent, width_scale, length_scale, rotation, collector, inheritance):
        
        # Store the parent segment, inherited parameter information, and the full segment list
        self.parent = parent
        self.inheritance = inheritance
        self.collector = collector
        
        # Add self to the collector
        self.idx = len(collector)
        self.collector.append(self)
        
        # Store the relative specification
        self.length_scale = length_scale
        self.width_scale = width_scale
        self.rotation = rotation
        
        if parent is None:
            
            self.length = inheritance["root"]["length"]
            self.width  = inheritance["root"]["width"]
            self.angle  = inheritance["root"]["angle"]
            self.startx = inheritance["root"]["startx"]
            self.starty = inheritance["root"]["starty"]
            
        else:
            
            # Calculate the absolute properties
            self.length = parent.length * self.length_scale
            self.width  = parent.width * self.width_scale
            self.angle  = (parent.angle + self.rotation) % (2*np.pi)
            self.startx = parent.startx + np.sin(parent.angle)
            self.starty = parent.starty + np.cos(parent.angle)
        
        #print("Created child {}".format(self.idx))
        self.spawn()
    
    
    @abstractmethod
    def spawn(self):
        pass
```

```python
class regular_segment(segment):
    """
    A segment which passes its relative specification deterministically to its children
    """
    
    def spawn(self):
        
        # Check that we're not too small
        if self.width > self.inheritance["width_tolerance"]:
            
            # Loop through children
            for ii in range(self.inheritance["order"]):
                
                # Create the child
                length_scale = self.inheritance["length_scale"][ii]
                width_scale = self.inheritance["width_scale"][ii]
                rotation = self.inheritance["rotation"][ii]
                child = regular_segment(self, width_scale, length_scale, rotation, self.collector, self.inheritance)
```

```python
inheritance = dict()
inheritance["width_tolerance"] = 1
inheritance["order"] = 2
inheritance["width_scale"] = [0.7]*2
inheritance["length_scale"] = [0.65,0.7]
inheritance["rotation"] = [0.6,-0.6]
inheritance["root"] = dict()
inheritance["root"]["length"] = 70
inheritance["root"]["width"] = 7
inheritance["root"]["angle"] = 0.0
inheritance["root"]["startx"] = 0.0
inheritance["root"]["starty"] = 0.0

collector = []
root = regular_segment(None, None, None, None, collector, inheritance)
print(len(collector))
```

```python
def draw_fractal_tree(collector, params):
    """
    Draws a fractal tree
    """
    
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    
    # Loop through the segments
    for segment in collector:
        
        # Draw it
        x1 = segment.startx
        y1 = segment.starty
        x2 = segment.startx + np.sin(segment.angle)
        y2 = segment.starty + np.cos(segment.angle)
        ax.plot([x1, x2], [y1, y2], color='g', linestyle='-', linewidth=segment.width)
        
    plt.show()
```
