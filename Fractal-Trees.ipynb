```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from abc import ABCMeta, abstractmethod 
```

# Abstract Fractal Tree Segment

```python
class segment:
    """
    A segment of a fractal tree
    """
    __metaclass__ = ABCMeta
    
    def __init__(self, parent, width_scale, length_scale, rotation, collector, inheritance):
        
        # Store the parent segment, inherited parameter information, and the full segment list
        self.parent = parent
        self.inheritance = inheritance
        self.collector = collector
        
        # Store the relative specification
        self.length_scale = length_scale
        self.width_scale = width_scale
        self.rotation = rotation
        
        if parent is None:
            
            # If we don't have a parent, get root properties
            self.length = inheritance["root"]["length"]
            self.width  = inheritance["root"]["width"]
            self.angle  = inheritance["root"]["angle"]
            self.startx = inheritance["root"]["startx"]
            self.starty = inheritance["root"]["starty"]
            
        else:
            
            # Calculate the absolute properties
            self.length = parent.length * self.length_scale
            self.width  = parent.width * self.width_scale
            self.angle  = (parent.angle + self.rotation) % (2*np.pi)
            self.startx = parent.startx + parent.length*np.sin(parent.angle)
            self.starty = parent.starty + parent.length*np.cos(parent.angle)
        
        # Stop if we go below ground
        if inheritance["kill_below_zero"] and ( self.starty + self.length*np.cos(self.angle) < 0.0 ):
            return
        
        # Add self to the collector
        self.idx = len(collector)
        self.collector.append(self)
        
        # Make baby segment
        self.spawn()
    
    
    @abstractmethod
    def spawn(self):
        pass
```

```python
def draw_fractal_tree(collector, params):
    """
    Draws a fractal tree
    """
    
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    ax.set_aspect("equal", adjustable="datalim")
    
    # Loop through the segments
    for segment in collector:
        
        # Draw it
        x1 = segment.startx
        y1 = segment.starty
        x2 = segment.startx + segment.length*np.sin(segment.angle)
        y2 = segment.starty + segment.length*np.cos(segment.angle)
        ax.plot([x1, x2], [y1, y2], color='g', linestyle='-', linewidth=segment.width)
        
    plt.show()
```

# Implemented Deterministic Segment

```python
class regular_segment(segment):
    """
    A segment which passes creates its children according to an inherited deterministic specification
    """
    
    def spawn(self):
        
        # Check that we're not too small
        if self.width > self.inheritance["width_tolerance"]:
            
            # Loop through children
            for ii in range(self.inheritance["order"]):
                
                # Create the child
                length_scale = self.inheritance["length_scale"][ii]
                width_scale = self.inheritance["width_scale"][ii]
                rotation = self.inheritance["rotation"][ii]
                child = regular_segment(self, width_scale, length_scale, rotation, self.collector, self.inheritance)
```

# Implemented Stochastic Segment

Symmetrical is nice, but the perfectly symmetric trees have an awkward blockiness. Maybe we should add some randomness...?

```python
class stochastic_segment(segment):
    """
    A segment which creates its children according to an inherited stochastic specification
    """
    
    def spawn(self):
        
        # Check that we're not too small
        if self.width > self.inheritance["width_tolerance"]:
            
            # Loop through children
            for ii in range(self.inheritance["order"]):
                
                # Create the child
                length_scale = np.random.normal(loc=self.inheritance["length_scale"][ii],
                                                scale=self.inheritance["scale_standard_deviation"])
                width_scale = self.inheritance["width_scale"][ii]
                rotation = np.random.normal(loc=self.inheritance["rotation"][ii],
                                            scale=self.inheritance["rotation_standard_deviation"])
                child = stochastic_segment(self, width_scale, length_scale, rotation, self.collector, self.inheritance)
```
